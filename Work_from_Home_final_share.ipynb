## Import Library

```
import pandas as pd
import gspread
from gspread_dataframe import get_as_dataframe, set_with_dataframe
```

### Read Data from excel 
ไฟล์ WFH Project Report.xlsx มี 2 sheet 
<br>
แต่ละsheet เป็นข้อมูลของแต่ละ project (project1_report,project2_report)


```
df_project_1=pd.read_excel("https://github.com/live2pro/dataset/raw/master/WFH%20Project%20Report.xlsx",sheet_name="project1_report")
df_project_2=pd.read_excel("https://github.com/live2pro/dataset/raw/master/WFH%20Project%20Report.xlsx",sheet_name="project2_report")
```

### Create dataframe 


1.   df_project_1 : dataframe ของ Project1
2.   df_project_2 : dataframe ของ Project2



```
df_project_1.head()
```

### Assign variable 

```
name_project1=df_project_1.columns[1]
name_project2=df_project_2.columns[1]
```

```
leader_project1=df_project_1.iloc[0,1]
leader_project2=df_project_2.iloc[0,1]
```

### Slice data to creat new DataFrame

```
df_project_1_clean=df_project_1.iloc[3:,]
df_project_2_clean=df_project_2.iloc[3:,]
```

```
df_project_1_clean
```

### Change column name

```
df_project_1_clean.columns=df_project_1.iloc[2,]
df_project_2_clean.columns=df_project_2.iloc[2,]
```

```
df_project_1_clean
```

### Create new columns "Project" and "Leader"

```
df_project_1_clean=df_project_1_clean.copy()
df_project_1_clean['Project']=name_project1
df_project_1_clean['Leader']=leader_project1
```

```
df_project_2_clean=df_project_2_clean.copy()
df_project_2_clean['Project']=name_project2
df_project_2_clean['Leader']=leader_project2
```

```
df_project_1_clean
```

```
df_project_2_clean
```

### Prepare data to transform wide to long format data (Unpivot)

```
#Convert column name to list
col_proj1=df_project_1_clean.columns.to_list()
col_proj2=df_project_2_clean.columns.to_list()
```

```
#Captue seletected range of values
date_p1=[col_proj1[i] for  i in range(3,15) ]
date_p2=[col_proj2[i] for  i in range(3,15) ]
```

```
#Unpivot : Transform wide to long format 
new_df_proj1=pd.melt(df_project_1_clean, id_vars=['Project','Leader','TaskId','Task','Response staff'], value_vars=date_p1,
        var_name='Date', value_name='Progress')
```

```
new_df_proj1
```

```
#Unpivot : Transform wide to long format 
new_df_proj2=pd.melt(df_project_2_clean, id_vars=['Project','Leader','TaskId','Task','Response staff'], 
                     value_vars=date_p2,var_name='Date', value_name='Progress')
```

```
new_df_proj2
```

### Clean dataset by removing null values

```
new_df_proj1.dropna(inplace=True)
new_df_proj2.dropna(inplace=True)
```

```
new_df_proj1.isnull().sum()
```

```
new_df_proj2.isnull().sum()
```

### Calculate Progress in %

```
# Convert Progress to Percent data
new_df_proj1['Progress_pct']=new_df_proj1['Progress']/100
new_df_proj2['Progress_pct']=new_df_proj2['Progress']/100
```

### Calculate each task to  % progress of its own project 

```
# แต่ละ task คิดเป็นกี่ % ของโปรเจ็ค
new_df_proj1['%Task_of_project']=new_df_proj1['Progress_pct']/(len(new_df_proj1['Task'].unique()))
new_df_proj2['%Task_of_project']=new_df_proj2['Progress_pct']/(len(new_df_proj2['Task'].unique()))
```

```
new_df_proj1.head(2)
```

```
new_df_proj2.head(2)
```

### Combine all dataframes to single dataframe

```
# Merge  project1 and project2 DataFrame into one
two_project=pd.concat([new_df_proj1,new_df_proj2])
```

```
# Create % Progress by  ความก้าวหน้าปัจจุบันคิดเป็นกี่% ของ task นั้นๆ
task_group=two_project.groupby('TaskId')
```

```
combine=[]
for k,v in task_group:
  combine.append(task_group.get_group(k))
```

```
# Create column Progress_by_task  เป็น cumdiff (ตรงข้ามกับ cumsum) 
for g in combine:
  g['Progress_by_task']=g['Progress_pct'].diff().fillna(g['Progress_pct'])
```
